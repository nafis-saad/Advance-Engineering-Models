```python
import requests
import json

# Base URL for the ArcGIS REST API endpoint
BASE_URL = "https://outagemap.serv.dteenergy.com/arcgis/rest/services/OMP/OutageLocations/MapServer"

# Parameters for the initial query
params = {
    'where': '1=1',                # Select all features
    'outFields': '*',              # Include all attributes
    'f': 'json',                   # Output format as JSON
    'resultOffset': 0,             # Pagination offset
    'resultRecordCount': 1000      # Number of records per batch
}

# List to hold all features
all_features = []

try:
    while True:
        # Make the API request
        response = requests.get(BASE_URL, params=params)
        response.raise_for_status()  # Raise an error for bad status codes

        # Parse the JSON response
        data = response.json()

        # Debug: Check if an error exists in the response
        if 'error' in data:
            raise Exception(f"API Error: {data['error']}")

        # Extract features
        features = data.get('features', [])
        if not features:
            break  # Stop if there are no more features to fetch

        # Append features to the list
        all_features.extend(features)
        print(f"Fetched {len(features)} records, total: {len(all_features)}")

        # Update resultOffset for the next batch
        params['resultOffset'] += params['resultRecordCount']

    # Save all features to a JSON file
    with open("outage_data.json", "w") as outfile:
        json.dump(all_features, outfile, indent=4)

    print("All data has been successfully downloaded and saved to 'outage_data.json'.")

except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import pandas as pd

# Assuming df is your DataFrame
# Calculate the number of missing values for each column
missing_values = gdf.isnull().sum()

# Calculate the percentage of missing values for each column
missing_percentage = (gdf.isnull().sum() / len(gdf)) * 100

# Combine into a single DataFrame for better visualization
missing_summary = pd.DataFrame({
    'Missing Values': missing_values,
    'Missing Percentage': missing_percentage
}).sort_values(by='Missing Values', ascending=False)

# Display the result
print(missing_summary)


```

```python
import pandas as pd
import json

# File path to the JSON file
file_path = "outage_data.geojson"

# Read the JSON file
with open(file_path, "r") as infile:
    geojson_data = json.load(infile)

# Extract features from the GeoJSON
features = geojson_data.get("features", [])

# Create a function to handle missing or nested data
def extract_properties(feature):
    properties = feature.get("properties", {})
    geometry = feature.get("geometry", {})
    # Extract properties and geometry details
    row = {**properties}
    row["geometry_type"] = geometry.get("type")
    row["coordinates"] = geometry.get("coordinates")
    return row

# Apply the function to all features
data = [extract_properties(feature) for feature in features]

# Convert the list of dictionaries to a pandas DataFrame
df = pd.DataFrame(data)

```
