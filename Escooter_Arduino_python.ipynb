```python
import serial
import time

try:
    # Open the serial port
    ser = serial.Serial('COM4', 9600, timeout=1)  # Adjust as per your Arduino's COM port

    # Open or create a file for logging
    with open("output.txt", "w") as file:
        while True:
            try:
                if ser.in_waiting > 0:
                    line = ser.readline().decode('utf-8').rstrip()  # Read a line from the serial port
                    file.write(line + "\n")  # Write the line to a file
                    file.flush()  # Ensure it gets written to disk
                    print(line)  # Optionally print the line
                    time.sleep(.01)  # Delay to match the Arduino timing
            except serial.SerialException as e:
                print(f"Error reading from the serial port: {e}")
                break  # Optionally break the loop on error
            except UnicodeDecodeError as e:
                print(f"Error decoding data: {e}")
                # Optionally continue or handle differently
            except Exception as e:
                print(f"Unexpected error: {e}")
                break  # Exit the loop on unexpected error

except serial.SerialException as e:
    print(f"Failed to open serial port: {e}")
except FileNotFoundError as e:
    print(f"Failed to open file: {e}")
except Exception as e:
    print(f"Unexpected error: {e}")
```
