```python
import pandas as pd
import os

# Directory containing the CSV files
directory_path = 'D:/Twitter_Project/Google_Review'

# List to store dataframes
dataframes = []

# Read each CSV file and append to the list
for filename in os.listdir(directory_path):
    if filename.endswith('.csv'):
        filepath = os.path.join(directory_path, filename)
        try:
            # Try reading with UTF-8
            df = pd.read_csv(filepath)
        except UnicodeDecodeError:
            # If UTF-8 fails, try another encoding, like ISO-8859-1
            df = pd.read_csv(filepath, encoding='ISO-8859-1')
        dataframes.append(df)

# Concatenate all dataframes row-wise
combined_df = pd.concat(dataframes, axis=0, ignore_index=True)
combined_df
```

```python
# Save the combined dataframe to a new CSV file
combined_df.to_csv('D:/Twitter_Project/Google_Review/combined_csv.csv', index=True)

print("CSV files have been combined successfully.")
```
