```python
import pandas as pd

# Load the dataset
file_path = 'D:/Flood_Project/Survey_dataset/Survey Formatted/final_research.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
df.head()
```

```python
# Clean up non-numeric values in the relevant columns, treating them as NaN if they cannot be converted
df['No_Vhcl_hhl'] = pd.to_numeric(df['No_Vhcl_hhl'], errors='coerce')
df['No_workingppl_hhl'] = pd.to_numeric(df['No_workingppl_hhl'], errors='coerce')

# Now re-apply the function to create the new column after cleaning up the data
def classify_vehicle_availability(row):
    vehicles = row['No_Vhcl_hhl']
    workers = row['No_workingppl_hhl']
    
    if pd.isna(vehicles) or pd.isna(workers):
        return 'Unknown'
    
    if vehicles == 0:
        return '0 Autos'
    elif 0 < vehicles < workers:
        return 'Autos > 0 & < Workers'
    elif vehicles >= workers:
        return 'Autos â‰¥ Workers'
    else:
        return 'Unknown'

# Apply the function again
df['Vehicle_Availability'] = df.apply(classify_vehicle_availability, axis=1)
```

```python
# Export the updated dataset to a CSV file
output_file_path = 'D:/Flood_Project/Survey_dataset/Survey Formatted/more_final_research.csv'
df.to_csv(output_file_path, index=False)


```
